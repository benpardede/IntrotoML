# **LABEL ENCODING**

```python
#Target Variables (Attrition)
df['Attrition'] = df['Attrition'].replace({'No':0, 'Yes':1})
```

```python
#encode binary variables
df['OverTime'] = df['OverTime'].map({'No':0, "Yes":1})
df['Gender'] = df['Gender'].map({'Male':0, "Female":1})
df['Over18'] = df['Over18'].map({'Y':1, "N":0})
```

```python
# encode categorical columns which are ordinal, use labelEncoding
# apply label encoder to df_categorical
from sklearn.preprocessing import LabelEncoder
```

```python
encoding_cols = ['BusinessTravel', 'Department','EducationField','JobRole', 'MaritalStatus'  ]
label_encoders = {}
for column in encoding_cols:
  label_encoders[column] = LabelEncoder()
  df[column] = label_encoders[column].fit_transform(df[column])
```

## <b><font color='white'> 2. RESAMPLING</font></b>
---

```python
X = df.drop(['Attrition'], axis =1) # Input Features
y = df['Attrition'].values #Output
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print((Counter(y)))
rus = RandomOverSampler(random_state = 42)
X_over, y_over = rus.fit_resample(X,y)
print(Counter(y_over))
```

```python
#Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size=0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
from sklearn.preprocessing import StandardScaler
# Buat scaler
scaler = StandardScaler()
scaler.fit(X)
```

```python
# Convert X_train & X_test
def scaler_transform(data, scaler):
    data_scaled = scaler.transform(data)
    data_scaled = pd.DataFrame(data_scaled)
    data_scaled.columns = data.columns
    data_scaled.index = data.index

    return data_scaled
```

```python
X_train_scaled = scaler_transform(data = X_train,
                                  scaler = scaler)
X_train_scaled.head()
```

```python
X_test_scaled = scaler_transform(data = X_test,
                                 scaler = scaler)
X_test_scaled.head()
```

## <b><font color='white'> 4. OBJECT FITTING</font></b>
---

**A. LOGISTIC REGRESSION**

```python
#Training model with Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train_scaled, y_train)
```

```python
prediction = logreg.predict(X_test_scaled)
conf_matrix = confusion_matrix(y_test, prediction)
```

```python
conf_matrix
```

```python
fig = plt.figure(figsize = (15,6))
ax1 = fig.add_subplot(1,2,1)
ax1 = sns.heatmap(pd.DataFrame(conf_matrix), annot = True, cmap = 'Blues', fmt = 'd')
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('Predicted')
plt.ylabel('Expected')

ax2 = fig.add_subplot(1,2,2)
y_pred_proba = logreg.predict_proba(X_test_scaled)[::,1]
fpr, tpr, _ = roc_curve(y_test, prediction)
auc = roc_auc_score(y_test, prediction)
ax2 = plt.plot(fpr,tpr, label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
auc
```

```python
#accuracy
from sklearn.metrics import accuracy_score
model_acc = accuracy_score(y_test, prediction)
print("Accuracy Score - ", model_acc)
```

```python
#Precision
from sklearn.metrics import precision_score
model_prec = precision_score(y_test, prediction)
print("Precision score of the model  - ", model_prec)
```

```python
#F1-Score
from sklearn.metrics import f1_score
f1score = f1_score(y_test, prediction)
print("F1-Score of the model - ", f1score)
```
